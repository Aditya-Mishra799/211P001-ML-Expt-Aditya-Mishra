```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('placement.csv')

df.head()
```

```python

# plotting the scatter plot
plt.scatter(df['cgpa'],df['package'])
plt.title('BY Aditya Mishra')
plt.xlabel('CGPA') # setting x label
plt.ylabel('Package(in lpa)') # setting y label
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X = df.iloc[:,0] # Getting cgpa values as X
Y = df.iloc[:,1] # Getting package values as Y
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0) # Splitting the data with test ratio as 0.2
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression() # Creating the model
model.fit(np.array(X_train).reshape(-1,1),Y_train) # Fitting the model
model.score(np.array(X_test).reshape(-1,1),Y_test) # Getting the score
print('Input CGPA : 7')
print('Predicted Package (in LPA) :- ')
print(model.predict(np.array(7).reshape(-1, 1))[0]) # Predicting the values
```

```python

plt.scatter(df['cgpa'],df['package'])
plt.plot(X_train,model.predict(np.array(X_train).reshape(-1,1)),color='orange')
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
plt.title('Vizulaization By Aditya Mishra')
plt.show()
```
