```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
# Load, standardize, and perform PCA on the Iris dataset
iris = load_iris()
iris_pca = PCA(n_components=2).fit_transform(StandardScaler().fit_transform(iris.data))
# Plot the PCA result
plt.figure(figsize=(8, 6))
plt.scatter(iris_pca[:, 0], iris_pca[:, 1], c=iris.target, cmap='viridis', edgecolor='k', s=100)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset By Aditya Shyamanand Mishra 211P001')

# Manually create the legend
for target, target_name in zip(np.unique(iris.target), iris.target_names):
    plt.scatter([], [], c=[plt.cm.viridis(target / 2)], label=target_name, edgecolor='k', s=100)
plt.legend(title="Species")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D

# Load, standardize, and apply PCA on Iris dataset
iris = load_iris()
iris_pca = PCA(n_components=3).fit_transform(StandardScaler().fit_transform(iris.data))

# Plot cumulative explained variance
explained_variance = np.cumsum(PCA(n_components=3).fit(StandardScaler().fit_transform(iris.data)).explained_variance_ratio_)
plt.plot(range(1, 4), explained_variance, marker='o', linestyle='--', color='b')
plt.title('Cumulative Explained Variance by Principal Components\nAditya Shyamanand Mishra 211P001')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(), plt.show()

# 2D PCA visualization
plt.scatter(iris_pca[:, 0], iris_pca[:, 1], c=iris.target, cmap='viridis', edgecolor='k', s=100)
plt.xlabel('Principal Component 1'), plt.ylabel('Principal Component 2')
plt.title('2D PCA of Iris Dataset By Aditya Shyamanand Mishra 211P001')
for i, name in enumerate(iris.target_names):
    plt.scatter([], [], c=[plt.cm.viridis(i / 2)], label=name, edgecolor='k', s=100)
plt.legend(title="Species"), plt.grid(), plt.show()

# 3D PCA visualization
fig = plt.figure(); ax = fig.add_subplot(111, projection='3d')
ax.scatter(iris_pca[:, 0], iris_pca[:, 1], iris_pca[:, 2], c=iris.target, cmap='viridis', edgecolor='k', s=100)
ax.set_xlabel('Principal Component 1'), ax.set_ylabel('Principal Component 2'), ax.set_zlabel('Principal Component 3')
ax.set_title('3D PCA of Iris Dataset By Aditya Shyamanand Mishra 211P001')
for i, name in enumerate(iris.target_names):
    ax.scatter([], [], [], c=[plt.cm.viridis(i / 2)], label=name, edgecolor='k', s=100)
ax.legend(title="Species"), plt.show()

# Print explained variance ratio
print("Explained variance ratio for each principal component:", explained_variance)
```
